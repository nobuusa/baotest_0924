```python
import json
import jieba
```

```python
data = json.load(open('routes/data.json'))
```

```python
data[0]
```

```python
from collections import Counter
cnt = Counter()
for i in range(len(data)):
    segs = [w for w in jieba.cut(data[i]['Q']) if len(w)>1]
    cnt += Counter(segs)
    data[i]['Keywords'] = segs#list(set(segs))
```

```python
cnt_valid = {}
for w, c in cnt.most_common():
    if len(w)<2 or c<30:
        continue
    cnt_valid[w] = c
```

```python
with open('routes/data_v2.json', 'w') as f:
    json.dump(data, f)
with open('routes/cnt.json', 'w') as f:
    json.dump(cnt_valid, f)
```
